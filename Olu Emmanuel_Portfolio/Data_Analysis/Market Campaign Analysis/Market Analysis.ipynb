```python
import pandas as pd 
bank = pd.read_csv('bank-full.csv', sep=';')
print('Data shape:',bank.shape)
print(bank.head())
print(bank.info())
print(bank.isna().sum())
bank['y'].value_counts(normalize = True )
conv_rate  = (bank['y'] == 'yes').mean()
print('Conversion Rate :',conv_rate * 100)
print(bank['y'].unique())

bank['converted'] = (bank['y'] == 'yes').astype(int)

bank[['age', 'campaign', 'duration']].describe()


def campaign_group(x):
    if x == 1:
        return '1 Contact'
    elif x <= 3:
        return '2–3 Contacts'
    else:
        return '4+ Contacts'

bank['campaign_intensity'] = bank['campaign'].apply(campaign_group)

print('Contact Channel Segmentation:', bank['contact'].value_counts(normalize=True) * 100)

print(bank['month'].value_counts())

bank[['campaign', 'campaign_intensity', 'contact', 'month']].head()

print('Campaign Intensity Counts:', bank['campaign_intensity'].value_counts())
print('Contact Method Percentage Split:', bank['contact'].value_counts(normalize=True) * 100)
print('conversion rate by campaign intensity',bank.groupby('campaign_intensity')['converted'].mean() * 100)
cr = bank.groupby('campaign_intensity')['converted'].mean() * 100
uplift = (cr.max() - cr.min()) / cr.min() * 100

print('uplift:', uplift)

print('Conversion Rate by Contact Method',bank.groupby('contact')['converted'].mean() * 100)

roi_df = bank.groupby('campaign_intensity').agg(
    total_customers=('campaign', 'count'),
    total_conversions=('converted', 'sum')
)

# Revenue & Cost
roi_df['revenue'] = roi_df['total_conversions'] * 50
roi_df['cost'] = roi_df['total_customers'] * 1

# ROI
roi_df['ROI'] = (roi_df['revenue'] - roi_df['cost']) / roi_df['cost'] * 100

print('ROI', roi_df)

roi_contact = bank.groupby('contact').agg(
    total_contacts=('contact', 'count'),  # total customers contacted
    total_conversions=('converted', 'sum')  # total converted
)


roi_contact['revenue'] = roi_contact['total_conversions'] * 50
roi_contact['cost'] = roi_contact['total_contacts'] * 1


roi_contact['ROI'] = (roi_contact['revenue'] - roi_contact['cost']) / roi_contact['cost'] * 100

print('ROI_CONTACT :',roi_contact)

bank.to_csv("marketing_campaign_clean.csv", index=False)
```
