```python
# Import Required Packages
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Loading and examining the dataset
p_df = pd.read_csv("penguins.csv")
p_df.head()

# Convert categorical variables to dummy/indicator variables
p_df_encoded = pd.get_dummies(p_df, drop_first=True)

# Standardizing the features
Sc = StandardScaler()
p_df_scaled = Sc.fit_transform(p_df_encoded)

inert_list = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(p_df_scaled)
    inert_list.append(kmeans.inertia_)

# Use range(1, 10) for x-axis to match inert_list length
plt.bar(range(1, 10), inert_list)
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()



kmeans = KMeans(n_clusters=3, random_state=42).fit(p_df_scaled)
labels = kmeans.labels_

plt.scatter(
    p_df['culmen_length_mm'], p_df['culmen_depth_mm'], c=labels,)
plt.xlabel('Culmen Length (mm)')
plt.ylabel('Culmen Depth (mm)')
plt.title('KMeans Clustering of Penguins')
plt.show()


numeric_columns = p_df_encoded.select_dtypes(include='number').columns.tolist()


p_df_encoded['cluster'] = kmeans.labels_

# Group by 'cluster' and calculate the mean of numeric columns
stat_penguins = p_df_encoded.groupby('cluster')[numeric_columns].mean()

# Display the resulting DataFrame
stat_penguins
```
